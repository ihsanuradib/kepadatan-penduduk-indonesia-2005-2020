```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import matplotlib.animation as animation
```

```python
# Load images
img1 = mpimg.imread("Peta Kepadatan 2005.png")
img2 = mpimg.imread("Peta Kepadatan 2010.png")
img3 = mpimg.imread("Peta Kepadatan 2015.png")
img4 = mpimg.imread("Peta Kepadatan 2020.png")
```

```python
# Create a list of images
images = [img1, img2, img3, img4]
```

```python
# Initialize the plot
fig, ax = plt.subplots()
im = ax.imshow(images[0])
```

```python
# Define the animation function
def animate(i):
    im.set_array(images[i % len(images)])
    return im,
```

```python
# Set the animation
ani = animation.FuncAnimation(fig, animate, frames=len(images), blit=True)
```

```python
# Show the plot
plt.show()
```

```python
ani.save('animation.gif', writer='imagemagick')
```
